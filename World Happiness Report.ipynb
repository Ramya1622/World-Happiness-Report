# Linear Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X=world_data_2015[['Family']]
#X = world_data_2015[['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity']]
```

```python
y = world_data_2015['Happiness Score']
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
#training linear regression model using available datapoints-fit
lm.fit(X,y)
```

```python
#intercept value-i
print(lm.intercept_)
```

```python
#coef value-c
lm.coef_
```

```python
lm.predict([[1.63]])
```

```python
#cdf = pd.DataFrame(lm.coef_,X.columns,columns=['Coeff'])
```

```python
predictions = lm.predict(X)
```

```python
plt.scatter(y,predictions)
```

```python
sns.displot(y-predictions)
```

```python
sns.set_style('white')
sns.color_palette("bright")
plt.figure(figsize=(20,8))
plt.title('Plot btw Family and Happiness Score',fontsize=15,weight='bold')
plt.xticks(fontsize=12,weight='bold')
plt.xlabel('Family',fontsize=12,weight='bold')
plt.yticks(fontsize=12,weight='bold')
plt.ylabel('Happiness Score',fontsize=12,weight='bold')
plt.scatter(data=world_data_2015, # define data source
           x='Family',             # column names
           y='Happiness Score')
plt.plot(world_data_2015.Family,lm.predict(X))
plt.show()
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y,predictions)
```

```python
metrics.mean_squared_error(y,predictions)
```

```python
np.sqrt(metrics.mean_squared_error(y,predictions))
```

## Relationship btw GDP & Happiness

```python
z=world_data_2015['Region']
sns.set_style('white')
sns.color_palette("bright")
plt.figure(figsize=(20,8))
plt.title('Plot btw GDP and Happiness Score',fontsize=15,weight='bold')
plt.xticks(fontsize=12,weight='bold')
plt.xlabel('Happiness Score',fontsize=12,weight='bold')
plt.yticks(fontsize=12,weight='bold')
plt.ylabel('GDP',fontsize=12,weight='bold')
sns.scatterplot(data=world_data_2015, # define data source
           x='Happiness Score',             # column names
           y='Economy (GDP per Capita)',
           hue=z,
           alpha=0.8,
           s=200)
plt.legend(loc='best')
plt.show()
```

## Relationship btw Health & Happiness

```python
sns.set_style('white')
sns.color_palette("bright")
plt.figure(figsize=(20,8))
plt.title('Plot btw Health and Happiness Score',fontsize=15,weight='bold')
plt.xticks(fontsize=12,weight='bold')
plt.xlabel('Happiness Score',fontsize=12,weight='bold')
plt.yticks(fontsize=12,weight='bold')
plt.ylabel('Health',fontsize=12,weight='bold')
sns.scatterplot(data=world_data_2015, # define data source
           x='Happiness Score',             # column names
           y='Health (Life Expectancy)',
           hue='Region',
           alpha=0.9,
           s=200)
plt.legend(loc='best')
plt.show()
```

```python
lst_1=[['Economy (GDP per Capita)',world_data_2015['Economy (GDP per Capita)'].sum()],['Family',world_data_2015['Family'].sum()],['Health (Life Expectancy)',world_data_2015['Health (Life Expectancy)'].sum()],['Freedom',world_data_2015['Freedom'].sum()],['Trust (Government Corruption)',world_data_2015['Trust (Government Corruption)'].sum()],['Generosity',world_data_2015['Generosity'].sum()]]
lst_1
```

```python
pd_df=pd.DataFrame(lst_1,columns=['Happiness Contributors','Happiness Contribution'])
pd_df
```

```python
plt.figure(figsize=(15,8))
plt.pie(pd_df['Happiness Contribution'],
        labels=pd_df['Happiness Contributors'].values,
        autopct='%.2f%%',
        textprops={'size':'x-large',
                  'fontweight':'bold'})
plt.show()
```

```python
#On Analysing the Pie Chart, we are able to identify that the Family, Health and GDP are the 3 major factors that contributed to the Happiness Rate.
#The least contributor among the different factors is the Government Corruption-Trust.
```

```python
life_expectancy = world_data_2015.groupby('Region')['Health (Life Expectancy)'].sum()
life_expectancy
```

```python
plt.figure(figsize=(20,10))
plt.plot(life_expectancy,marker='d')
plt.xticks(rotation=30)
plt.show()
         
```

```python
#life expectancy is very low in North America whereas the highest has been marked in Central & Eastern Europe
```

```python
df_stack_plot=world_data_2015.groupby('Region').sum()
df_stack_plot['Region']=df_stack_plot.index
df_stack_plot
```

# BAR CHART- REGION WISE

```python
plt.figure(figsize = (15,6))
sns.set_style('white')
plt.bar(df_stack_plot['Region'],height=df_stack_plot['Happiness Score'], color = 'midnightblue')
plt.xticks(rotation = 45, fontsize = 10, fontweight = 'bold')
plt.yticks(fontsize = 10, fontweight = 'bold')
plt.title('Region Wise Happiness Measure',fontsize=20,weight='bold')
plt.show()
```

```python
#We can observe that Sub-Saharan Africa is the happiest region among the others. Also the Region with least happiness score is North America
```

```python
plt.figure(figsize=(20,8))
labels=('Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity')
plt.stackplot(df_stack_plot['Region'],
              df_stack_plot['Economy (GDP per Capita)'],
              df_stack_plot['Family'],
              df_stack_plot['Health (Life Expectancy)'],
              df_stack_plot['Freedom'],
              df_stack_plot['Trust (Government Corruption)'],
              df_stack_plot['Generosity'],
              edgecolor='none')
plt.title('Plot of Region & Happiness Factors',fontsize=20,weight='bold')
plt.xticks(rotation=30,fontsize=15,weight='bold')
plt.yticks(fontsize=15,weight='bold')
plt.ylabel('Happiness Contribution',fontsize=15,weight='bold')
plt.xlabel('Region',fontsize=15,weight='bold')
plt.legend(labels=labels,loc='best')
plt.show()
```
